```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
x = np.arange(0, 2*pi, 0.1)

y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.tanh(x)

fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y1)
axs[0,0].set_title("Sine Function")
axs[0,1].plot(x,y2)
axs[0,1].set_title("Cosine Function")
axs[1,0].plot(x,y3)
axs[1,0].set_title("Tangent Function")
axs[1,1].plot(x,y4)
axs[1,1].set_title("Tanh Function")

plt.show()
```
